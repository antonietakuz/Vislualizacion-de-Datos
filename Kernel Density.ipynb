### Kernel Density Estimation (KDE)

#### 1. Trabajamos con los datos de corredores

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
corredores =pd.read_csv('C:/Users/pc/Desktop/rociochavez/Rocio-Chavez-youtube-Files-master/categorias de corredores.csv')
```

```python
corredores.head()
```

```python
regular =corredores.loc[(corredores['Velocidad']=='Regular'), ['Velocidad', 'Edad']]
regular
```

#### 2. Creamos un histograma

```python
sns.histplot(data =regular['Edad'])
plt.title('Histograma de corredores')
plt.show()
```

#### 3. Grafico de densidad KDE

```python
plt.rcParams['figure.figsize']=(6,4.4)
sns.histplot(data=regular['Edad'], kde=True)
plt.show()
```

#### 4. Obtenemos solamente el grafico de densidad generado mediante el metodo KDE

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
sns.kdeplot(data=regular [ 'Edad'], color ='g', shade=True)
ax.set_ylim(0, 0.06)
ax.set_xlim(0, 100)
ax.set_title('Grafico de densidad corredores velocidad regular')
plt.show()
```

#### 5. Obtenemos el histograma con su grafico de densidad filtrando la informacion dentro del codigo

```python
data =corredores.loc[(corredores['Velocidad']=='Lento'), 'Edad']
data
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
sns.kdeplot(data, color='b', shade=True), 
ax.set_ylim(0, 0.06)
ax.set_xlim(0, 100)
ax.set_title('Grafico de densidad corredores velocidad regular')
plt.show()
```

#### 6. Desplegamos todas las categorias de la variable velocidad en un solo grafico sin necesidad de dividir informacion por categorias de velocidad

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)

sns.kdeplot(data=corredores, x='Edad', hue='Velocidad', shade=True), 
ax.set_ylim(0, 0.04)
ax.set_xlim(0, 100)

plt.show()
```

#### 7. Grafico de Densidad tomando en cuenta dos variables

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)

sns.kdeplot(data=corredores, y='Tiempo', x='Edad', hue='Velocidad', shade=True), 
ax.set_ylim(100, 450)
ax.set_xlim(10, 80)

plt.show()
```
