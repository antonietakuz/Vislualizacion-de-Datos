### Tablas pivote con Pandas

```python
#Obtenemos la suma de los valores de Costo por Kilo por tipo de hortaliza
hidroponia.pivot_table(index='Sistema', values=['Costo', 'Ganancia'], aggfunc= np.sum)
```

```python
#Le añadimos los sutotales y el total general

hidroponia.pivot_table(index='Sistema', values=['Costo', 'Ganancia'], aggfunc= np.sum, margins=True)
```

```python
#Obtenemos el promedio

hidroponia.pivot_table(index='Sistema', values=['Costo', 'Ganancia'], aggfunc=np.mean)
```

```python
#Incluimos dos variables en las filas de la tabla

hidroponia.pivot_table(index=['Sistema', 'Hortaliza'], values=['Costo'], aggfunc=np.sum)
```

```python
# Pedimos que nos desgloce los costos por cada valor 

hidroponia.pivot_table(index=['Sistema', 'Hortaliza'], values=['Costo'], columns = ['pH'], aggfunc=np.sum)
```

```python
# Reemplazamos los NaN con ceros

hidroponia.pivot_table(index=['Sistema', 'Hortaliza'], values=['Costo'], columns = ['pH'], aggfunc=np.sum, fill_value=0)
```

```python
#Si queremos agregar mas funciones, le pasamos los valores de aggfunc por medio de diccionario

np.round(hidroponia.pivot_table(index= 'Hortaliza', values=['Costo'], columns = ['Sistema'], aggfunc={'Costo':[np.mean, np.sum], 'Sistema':len}, fill_value=0),2)
```
