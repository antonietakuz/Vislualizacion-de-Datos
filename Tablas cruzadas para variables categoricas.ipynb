### Tablas cruzadas
Para variables categoricas y necesitamos saber cuantos individuos presentan

```python
#tablas cruzadas con pandas

pd.crosstab(hidroponia.Hortaliza, hidroponia.Sistema)
```

```python
#tablas cruzadas con pandas y totales

pd.crosstab(hidroponia.Hortaliza, hidroponia.Sistema, margins=True)
```

```python
#tablas cruzadas con pandas y porcentajes

pd.crosstab(hidroponia.Hortaliza, hidroponia.Sistema, margins=True, normalize=True)
```

```python
#Pocentaje por columnas
pd.crosstab(hidroponia.Hortaliza, hidroponia.Sistema, normalize='columns')
```

```python
pd.crosstab([hidroponia.Sistema, hidroponia.Hortaliza], hidroponia.pH)
```
