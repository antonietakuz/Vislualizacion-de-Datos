## Graficas de tipo violin

#### 1. Trabajamos con los datos de corredores

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
corredores =pd.read_csv('C:/Users/pc/Desktop/rociochavez/Rocio-Chavez-youtube-Files-master/categorias de corredores.csv', index_col=0)
```

```python
corredores.head()
```

#### 2. Creacion grafica Violin

```python
violin=sns.catplot(x='Velocidad', y='Edad', data=corredores.loc[(corredores['Velocidad']=='Regular'), ['Velocidad', 'Edad']], kind='violin', color='g')
violin.set_xticklabels(rotation=45)
plt.show()
```

#### 3. Obtenemos los graficos de violin de las tres categorias de las variables velocidad

```python
violin=sns.catplot(x='Velocidad', y='Edad', data=corredores, kind='violin', palette='Paired')
violin.set_xticklabels(rotation=45)
plt.show()
```

#### 4. Desglozamos la información por género

```python
violin=sns.catplot(x='Edad', y='Velocidad', data=corredores, kind='violin', palette='Dark2', col='Genero')
violin.set_xticklabels(rotation=45)
plt.show()
```

```python
violin=sns.catplot(x='Edad', y='Velocidad', data=corredores, kind='violin', palette='Dark2', hue='Genero')
violin.set_xticklabels(rotation=45)
plt.show()
```

```python
#grafica de genero por cada una de las velocidades

violin=sns.catplot(x='Velocidad', y='Edad', data=corredores, kind='violin', palette='Dark2', hue='Genero', split=True)
violin.set_xticklabels(rotation=45)
plt.show()
```
